## Financial Fraud Exploratory Data Analysis

The purpose of this notebook is to perform some initial exploratory data analysis to get a better understanding of our dataset. This will include univariate, bivariate and multivariate analysis. This analysis will help form some thoughts on how to approach an ensemble classifier in future notebooks. 

### Imports

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#loading in our data set
df = pd.read_csv("https://s3.amazonaws.com/dstkh.datasets/PS_20174392719_1491204439457_log.csv")
```

```python
#get an overall look at our data
df.info()
```

```python
df.head()
```

### Analysis & Visualizations

With an initial look at our dataset it seems we have a mixture of categorical and numerical data. It could be helpful to begin with some univariate analysis to get a better understanding of our columns by themselves. 

```python
#plot distribution of type column
sns.countplot(x='type', data=df)
```

```python
#plot distribution of isFraud column
sns.countplot(x='isFraud', data=df)
```

```python
#get numerical counts
df['isFraud'].value_counts()
```

```python
#plot distribution of isFlaggedFraud column
sns.countplot(x='isFlaggedFraud', data=df)
```

```python
#get numerical counts
df['isFlaggedFraud'].value_counts()
```

```python
#plot distribution of amount column
plt.figure(figsize=(10, 10))
sns.displot(x=df['amount'], kde=True)
```

Lets see how our data is affected by the type of transaction that is taking place.

```python
#boxplots of our amount by type
sns.boxplot(data=df, x='amount', y='type')
```

```python
#boxplots of old and new balance origin by type
fig, ax = plt.subplots(1, 2, figsize=(12.5, 10))
sns.boxplot(data=df, x='oldbalanceOrg', y='type', ax=ax[0])
sns.boxplot(data=df, x='newbalanceOrig', y='type', ax=ax[1])
```

```python
#boxplots of old and new balance destination by type
fig, ax = plt.subplots(1, 2, figsize=(12.5, 10))
sns.boxplot(data=df, x='oldbalanceDest', y='type', ax=ax[0])
sns.boxplot(data=df, x='newbalanceDest', y='type', ax=ax[1])
```

```python
#scatter plot of old and new balance from origin by isFraud
sns.scatterplot(data=df, x='oldbalanceOrg', y='newbalanceOrig', hue='isFraud')
```

```python
#scatter plot of old and new balance of destination by isFraud
sns.scatterplot(data=df, x='oldbalanceDest', y='newbalanceDest', hue='isFraud')
```

```python
#create a correlation heatmap of our numerical columns
plt.figure(figsize=(12.5, 10))
sns.heatmap(df.corr(numeric_only=True), annot=True)
```

### Takeaways

As we can see visualizing this dataset is not easy due to the presence of many outliers that make it hard to visualize a distribution. Although outliers make it hard to visualize our data it can be beneficial to our model in the future since it could make sense that fraudulent transactions are represented by abnormal quantities. 

Some key take aways that are important to note from this dataset is the fact that non-fraudulent transactions greatly outweigh fraudulent transactions. We also notice that `Payment`, `Cash Out` and `Cash In` are the more common types of transactions. 

Finally our scatter plots gave us a key insight as to how fraudulent and non-fraudelent transactions are distributed. From our `oldbalanceDest` and `newbalanceDest` scatter plot we can see that there is a shift between fraudulent transactions and non-fraudulent, this can be a key insight when building our model. There are also some columns that are correlated with each other such as the `oldbalanceOrg` and `newbalanceOrig` columns.
